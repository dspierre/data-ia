```python
data = {
    'Title': [],
    'Rating': [],
    'Price': [],
    'Available':[]}
print(data)
df = pd.DataFrame(data, columns=['Title', 'Rating', 'Price','Available'])
df.tail()
def text_to_number(letter_text):
    number_text = list(range(6))
    if letter_text=="Zero":
        return number_text[0]
    elif letter_text=="One":
        return number_text[1]
    elif letter_text=="Two":
        return number_text[2]
    elif letter_text=="Three":
        return number_text[3]
    elif letter_text=="Four":
        return number_text[4]
    elif letter_text=="Five":
        return number_text[5]
titre = [t.text for t in library_soup.findAll('h3')]
nota = [text_to_number(n.get('class')[1]) for n in library_soup.select('p.star-rating')] #int
prix = [p.text[1:] for p in library_soup.select('p.price_color')]
disp = [d.get('class')[0]=='instock' for d in library_soup.select('p.availability')]
df1000 = pd.DataFrame({'titre':titre,'notation':nota,'prix':prix,'disponibilité':disp})
df1000.head()
```

```python
end_page_num = 50
titre,nota,prix,disp = [],[],[],[]
for i in range(end_page_num+1):
    page_content = requests.get("http://books.toscrape.com/catalogue/page-{}.html".format(i)).content
    library_soup = BeautifulSoup(page_content, 'html.parser')
    titre += [t.text for t in library_soup.findAll('h3')]
    nota += [text_to_number(n.get('class')[1]) for n in library_soup.select('p.star-rating')] #int
    prix += [p.text[1:] for p in library_soup.select('p.price_color')]
    disp += [d.get('class')[0]=='instock' for d in library_soup.select('p.availability')]
df1000 = pd.DataFrame({'titre':titre,'notation':nota,'prix':prix,'disponibilité':disp})
print(df1000.shape,'\n',df1000.head())
```
