**4. Créer la matrice `A` suivante comme une concaténation de plusieurs vecteurs.**


$$A = \begin{bmatrix} 15 & -11 & 5 \\ 20 & -15 & 8 \\ 8 & -7 & 6 \end{bmatrix}$$



```python
a1 = [[15,-11,5]]
# liste
#a1 = [[15,-11,5]]
# liste (de liste)
#a1 = ([[15,-11,5]])
# tuple de listes
```

```python
a2 = [[20,-15,8]]
a3 = [[8,-7,6]]
# a2 = np.array([[20,-15,8]])
```

```python
A = np.concatenate((a1,a2,a3),axis=0) # vecteur = une ligne si axis= none => une ligne si axis = 1 => une ligne
print(A)
```

```python
# test des paramètres
A = np.concatenate((a1,a2,a3),axis=None)
print(A)
```

```python
# autre méthode
A = np.vstack((a1,a2,a3))
print(A)
```

```python
# par les colonnes 
c1 = ([15,20,8])
c2 = (-11,-15,-7)
c3 = (5,8,6)
A = np.column_stack((c1,c2,c3))
print(A)
```

**8. Extraire la diagonale de `A` dans une variable appelée `d`. De quel type est la variable ainsi créée?**

```python
print(A)
d = np.diag(A,k=0)
print(d)
```

```python
# Autre méthode 
d = A.diagonal()
print(d)
```

**11. Créer la matrice 
$$Q = \begin{bmatrix} 1 & 0\\ 0 & -1\end{bmatrix}$$**


```python
Q = np.array([[1,0],[0,-1]])
print(Q)
```

**12. Calculer sa transposée notée `T` et son inverse `V`à l'aide de la librairie Numpy. Que remarquez-vous? Que peut-on en déduire sur la nature de cette matrice?**


```python
from numpy import array
from numpy.linalg import inv
```

```python
T = np.transpose(Q)
print(T)
```

```python
# Autre méthode
T = Q.T
print(T)
```

```python
V = inv(Q)
print(V)
```

```python
# Autre méthode
V = np.linalg.inv(Q)
print(V)
```
