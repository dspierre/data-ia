Stemming

Le processus de « lemmatisation » consiste à représenter les mots (ou « lemmes ») sous leur forme canonique. Par exemple pour un verbe, ce sera son infinitif. Pour un nom, son masculin singulier. L'idée étant encore une fois de ne conserver que le sens des mots utilisés dans le corpus. Si l'on reprend notre exemple précédent, "Bonjour, je suis un texte d'exemple pour d'occurrences du verbe être plutôt que de compter séparément chaque usage de conjugaison de ce même verbe. De même pour les pluriels etc.

K-moyennes

Le partitionnement en k-moyennes (ou k-means en anglais) est une méthode de partitionnement de données et un problème d'optimisation combinatoire. Étant donnés des points et un entier k, le problème est de diviser les points en k groupes, souvent appelés clusters, de façon à minimiser une certaine fonction. On considère la distance d'un point à la moyenne des points de son cluster ; la fonction à minimiser est la somme des carrés de ces distances. Il existe une heuristique classique pour ce problème, souvent appelée méthodes des k-moyennes, utilisée pour la plupart des applications. Le problème est aussi étudié comme un problème d'optimisation classique, avec par exemple des algorithmes d'approximation.

Les k-moyennes sont notamment utilisées en apprentissage non supervisé où l'on divise des observations en k partitions. Les nuées dynamiques sont une généralisation de ce principe, pour laquelle chaque partition est représentée par un noyau pouvant être plus complexe qu'une moyenne. Un algorithme classique de k-means est le même que l'algorithme de quantification de Lloyd-Max.
inconvenient et aventage

Avantages et inconvénients pour l'apprentissage Un inconvénient possible des k-moyennes pour le partitionnement est que les clusters dépendent de l'initialisation et de la distance choisie[réf. souhaitée]. Le fait de devoir choisir a priori le paramètre k peut être perçu comme un inconvénient ou un avantage. Dans le cas du calcul des sac de mots par exemple, cela permet de fixer exactement la taille du dictionnaire désiré. Au contraire, dans certains partitionnements de données, on préférera s'affranchir d'une telle contrainte.

## 1. Scrapping

```python
titre,genre,an,synopsis,notenum = [],[],[],[],[]
page_content = requests.get('https://www.imdb.com/list/ls091520106/').content
library_soup = BeautifulSoup(page_content, 'html.parser')
titre = [t.text for t in library_soup.select('h3 a[href]')]
genre = [g.text for g in library_soup.select('span.genre')]
an = [int(a.text[-5:-1]) for a in library_soup.select('span.lister-item-year')]
synopsis = [s.text for s in library_soup.find_all('p', class_='')][1:]
notenum = [float(n.text) for n in library_soup.select('span.ipl-rating-star__rating') if len(n.text)==3]
notem=round(np.mean([m for m in notenum]),1)
notenum.append(notem)
notenum.append(notem)
df100 = pd.DataFrame({'titre':titre,'genre':genre,'année':an,'synopsis':synopsis,'note':notenum})
print(df100.shape,'\n',df100.head())
```

##### 2.4.2.1. TfidfVectorizer : tokenize + stem + stop words

```python
punc = ['.', ',', '"', "'", '?', '!', ':', ';', '(', ')', '[', ']', '{', '}',"%"]
stop_words = text.ENGLISH_STOP_WORDS.union(punc)

stemmer = SnowballStemmer('english')
tokenizer = RegexpTokenizer(r'[a-zA-Z\']+')

def tokenize(t):
    return [stemmer.stem(word) for word in tokenizer.tokenize(t.lower())]

vectorizer = TfidfVectorizer(strip_accents = 'unicode', stop_words = stop_words, tokenizer = tokenize, max_features = 100)
syn = vectorizer.fit_transform(df100.synopsis)
feat = vectorizer.get_feature_names()

print(syn.shape, '\n',feat[:50])
```

## 3. Classification non supervisée K-Means

Le score silhouette varie de -1 à 1. -1 pire valeur, 0 clusters se chevauchant, 1 meilleure valeur.

```python
range_n_clusters = [2,5,10,20,30]

for n_clusters in range_n_clusters:
    # Diviser en subplots
    fig, (ax1, ax2) = plt.subplots(1, 2)
    fig.set_size_inches(18, 7)

    # 1er subplot : silhouette
    ax1.set_xlim([-0.1, 1])
    # (n_clusters+1)*10 pour espacement silhouettes, plots individuels des clusters pour séparation
    ax1.set_ylim([0, syn.shape[0] + (n_clusters + 1) * 10])

    # KMeans
    clusterer = KMeans(n_clusters=n_clusters, random_state=10)
    cluster_labels = clusterer.fit_predict(syn)

    silhouette_avg = silhouette_score(syn, cluster_labels)
    print("Pour n_clusters =", n_clusters,
          "Silhouette score moyen =", silhouette_avg)

    # Déterminer les score silhouette
    sample_silhouette_values = silhouette_samples(syn, cluster_labels)

    y_lower = 10
    for i in range(n_clusters):
        # Assembler les scores silhouette pour les samples d'un même cluster i, puis les trier
        ith_cluster_silhouette_values = \
            sample_silhouette_values[cluster_labels == i]

        ith_cluster_silhouette_values.sort()

        size_cluster_i = ith_cluster_silhouette_values.shape[0]
        y_upper = y_lower + size_cluster_i

        color = cm.nipy_spectral(float(i) / n_clusters)
        ax1.fill_betweenx(np.arange(y_lower, y_upper),
                          0, ith_cluster_silhouette_values,
                          facecolor=color, edgecolor=color, alpha=0.7)

        # Label silhouette avec n° des clusters
        ax1.text(-0.05, y_lower + 0.5 * size_cluster_i, str(i))

        # Plot suivant
        y_lower = y_upper + 10  # 10 for the 0 samples

    ax1.set_title("Silhouette plot")
    ax1.set_xlabel("Coefficients")
    ax1.set_ylabel("Cluster n°")

    # Ligne moyenne des silhouette scores
    ax1.axvline(x=silhouette_avg, color="red", linestyle="--")

    ax1.set_yticks([])
    ax1.set_xticks([-0.1, 0, 0.2, 0.4, 0.6, 0.8, 1])

    # Clusters
    colors = list(cm.nipy_spectral(cluster_labels.astype(float) / n_clusters))
    ax2.scatter(syn[:, 0].toarray(), syn[:, 1].toarray(), marker='.', s=30, lw=0, alpha=0.7, c=colors, edgecolor='k')

    # Labels
    centers = clusterer.cluster_centers_
    # Cercles aux centres des clusters
    ax2.scatter(centers[:, 0], centers[:, 1], marker='o', c="white", alpha=1, s=200, edgecolor='k')

    for i, c in enumerate(centers):
        ax2.scatter(c[0], c[1], marker='$%d$' % i, alpha=1, s=50, edgecolor='k')

    ax2.set_title("Clusters")
    ax2.set_xlabel("1er feature")
    ax2.set_ylabel("2e feature")

    plt.suptitle(("Silhouette scores pour KMeans n_clusters = %d" % n_clusters),fontsize=14, fontweight='bold')

plt.show()
```
